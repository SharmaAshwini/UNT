```python
'''
Encode the categorical columns to numeric. There are two types of encoding: ordinal and one-hot. Explain why you choose the encoding technique to the column(s) and implement it. Show some rows of df after encoding. There will be no printed console in this subtask
Reference (you may need incognito mode to browse the pages):
    https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OrdinalEncoder.html
    https://towardsdatascience.com/guide-to-encoding-categorical-features-using-scikit-learn-for-machine-learning-5048997a5c79
    https://stackoverflow.com/questions/56502864/using-ordinalencoder-to-transform-categorical-values
    https://stackoverflow.com/questions/37292872/how-can-i-one-hot-encode-in-python
    https://pandas.pydata.org/docs/reference/api/pandas.get_dummies.html
'''
```

```python
df.info()
```

```python
#ordinal encoding for 'card type'
from sklearn.preprocessing import OrdinalEncoder

one_hot = df.copy()
categories = [['Signature', 'Silver', 'Gold', 'Platinum']]
enc = OrdinalEncoder(categories=categories)
one_hot['card type'] = enc.fit_transform(one_hot[['card type']])

one_hot.head(5)
```

```python
#inverse transform
enc.inverse_transform([[1]])
```

```python
#one-hot encoding for rest of the columns
one_hot = pd.get_dummies(one_hot, prefix=['city', 'exp type', 'gender'])

one_hot.head(5)
```

```python
#new encoded columns
one_hot.info()
```

**Observation:**

We have used both one-hot and ordinal encoding for this exercise.

Ordinal encoding has been used for 'card type' while One-hot has been used for rest of the columns.

**Why Ordinal encoding:**

The reason I chose Ordinal encoding for 'card type' feature, as order (ranking) is associated with this field. If we understand the domain, we can surely conclude that platinum card has more importance than gold or silver. Hence, there is some sort of order associated, and we can use ordinal encoding.


**Why One-hot encoding:**

The reason I chose one-hot encoding over ordinal for rest of the columns is the nature of data present in this dataset. Remaining categorical columns are city, 'exp type' and gender, which don't have any order or ranking associated. For e.g., there is no order associated in gender for female over male or viceversa. Hence, data type is Nominal for categorical columns and not Ordinal.


One-hot is suitable for Nominal category data, and panda's get_dummies implements it in a single line, efficiently.

```python
# Using sklearn library to split df into df_seen and df_unseen
```

```python
'''
Use KMeans method from sklearn to
    Fit df_seen
    Show the first five labels of df_unseen after prediction
'''
```

```python
from sklearn.model_selection import train_test_split


# Split the DataFrame into df_seen and df_unseen
df_seen, df_unseen = train_test_split(one_hot, test_size=0.3, random_state=0)

# Display the results
print("df_seen:")
df_seen.head(5)
```

```python
print("df_unseen:")
df_unseen.head(5)
```

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
#checking the number of clusters from a range (1 to 10) - Elbow method
distortions = []
for i in range(1, 11):
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(df_seen)
    distortions.append(km.inertia_)

# plot
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

Moving ahead with #clusters as 4

```python
from sklearn.cluster import KMeans
import numpy as np
np.random.seed = 100

kmeans = KMeans(n_clusters=4, init='random', random_state=0, n_init='auto', max_iter=300)

kmeans.fit(df_seen)

kmeans.labels_
```

```python
y_kmeans_unseen = kmeans.predict(df_unseen)
```

```python
y_kmeans_unseen[0:5]
```

```python
kmeans.inertia_
```

```python
#creating a new dataframe and storing assigned cluster numbers for future comparisons

df_uns_cluster = df_unseen.copy()
df_uns_cluster['cluster'] = y_kmeans_unseen
```

What should be the best number of clusters? Why? Is there a way to find it scientifically?

**Elbow method** can be used to find the number of clusters.

It is easy to understand as it provides optimum #clusters in a graphical manner. As we learnt in during lecture, if k increases, the Sum of Squared Errors(SSE) within cluster will decrease, as samples will be closer to assigned centroids.

We pick the value of k, where SSE begins to decrease most rapidly, mentioned in the next plot.

```python
#calculate distortion for a range of number of cluster
distortions = []
for i in range(1, 11):
    kmeans_clusters = KMeans(
        n_clusters=i, init='random',
        n_init='auto', max_iter=300,
        tol=1e-04, random_state=0
    )
    kmeans_clusters.fit(df_seen)
    distortions.append(kmeans_clusters.inertia_)

# plot
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

As we can see in the resulting plot, the elbow is located at k = 4, which is indeed a good choice for this dataset.
