##Train and Test Split

```python
import numpy as np
from sklearn.model_selection import train_test_split

X = new_df.iloc[:,1:].values

y = new_df.iloc[:,0].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

print('Training set:')
X_train
```

##Regression Function

```python
# Loss function MSE
def mse(y_true, y_pred):
  return np.mean((y_true - y_pred)**2)
```

```python
class My_LinearR() :

    def __init__( self, learning_rate, iterations ) :

        self.learning_rate = learning_rate

        self.iterations = iterations

    # Fit Function (model training)
    def fit( self, X, Y ) :

        self.m, self.n = X.shape        # samples and features

        self.W = np.zeros( self.n )     # initialize coefficient (weights)

        self.b = 0                      # initialize bias

        self.X = X                      # training columns

        self.Y = Y                      # result column


        # gradient descent learning
        for i in range( self.iterations ) :

            self.update_weights()   # tuning weights for mentioned epochs

        return self


    # Gradient descent implementation function (to update weights)
    def update_weights( self ) :

        Y_pred = self.predict( self.X )

        print(f'MSE: {mse(self.Y, Y_pred)}')

        grad_weight = ( self.X.T ).dot( -2*(self.Y - Y_pred)/self.m )   # calculate weight gradients

        grad_bias = np.sum( -2*(self.Y - Y_pred)/self.m )               # calculate bias gradient

        self.W = self.W - self.learning_rate * grad_weight              # update weights

        self.b = self.b - self.learning_rate * grad_bias                # update bias

        return self


    # Predict function
    def predict( self, X ) :

      ##print(f'X in predict: {X}')

      calc = X.dot( self.W ) + self.b

      return calc
```

```python
# Running the code for 1000 epochs / iterations
reg = My_LinearR(iterations = 1000, learning_rate = 0.01)
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
```
